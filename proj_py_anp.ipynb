```python
from jupyterthemes import jtplot
jtplot.style(theme = 'monokai')
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Vamos trazer os dados para 1° fazer uma primeura análise 

df = pd.read_csv('2023_01_producao_presal.csv',sep = ';')
df.head()
```

```python


df = df.loc[:,~df.columns.str.contains('^Unnamed')]
df.head()
```

```python
df.columns = df.columns.str.strip() 
```

```python
dados_df = df[["Tempo de Produção (hs por mês)","Petróleo (bbl/dia)"]]
dados_df
```

```python
df = df.rename(columns={"Tempo de Produção (hs por mês)": "Tempo_de_Producao", "Petróleo (bbl/dia)": "Petroleo"})
df.query("Tempo_de_Producao == 0 | Petroleo == 0").index
```

```python
dados_limpos = df.query("Tempo_de_Producao == 0 | Petroleo == 0").index
```

```python
df.drop(dados_limpos,axis= 0, inplace = True) 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecionar as duas colunas que você quer
df_limpo = df[['Tempo_de_Producao', 'Petroleo']]

# Criar o gráfico com seaborn
plt.figure(figsize=(8, 4))
sns.histplot(df_limpo['Tempo_de_Producao'], bins=20, kde=True, color='blue', label='Tempo de Produção', alpha=0.6)
sns.histplot(df_limpo['Petroleo'], bins=20, kde=True, color='green', label='Petróleo', alpha=0.6)

plt.title('Tempo de Produção e Petróleo')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.legend()
plt.show()


```
